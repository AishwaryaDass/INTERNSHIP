```
#observing relationship between different parameters

sns.barplot(x='Region',y='Profit',data=per)
```

```
plt.figure(figsize=(10,15))
ax=sns.barplot(x='Profit',y='State',data=per)
```

```
sns.barplot(x='Category',y='Profit',data=per)
```

```
sns.barplot(x='Profit',y='Sub-Category',data=per)
plt.show()
```

```
sns.jointplot(x='Category',y='Region',data=per)
plt.show()
```

```
sns.jointplot(x='Region',y='Sub-Category',data=per)
plt.figure(figsize=(10,6))
plt.show()
```

```
sns.jointplot(x='Category',y='State',data=per)
```

```
sns.jointplot(x='Sub-Category',y='State',data=per)
```

```
#the analysis for city vs profit and state with category and subcategory is not very clear as the data is per large.
#hence to establish a relationship between them we use the pair plot for easy analysis

sns.pairplot(per)
plt.show()
```

```
#trying to bring relationship between profit and discount

sns.barplot(x='Discount',y='Profit',data=per)
plt.show()
```

```
#trying to see with the profit varies wrt the mode of shipment
sns.barplot(x='Ship Mode',y='Profit',data=per)
plt.show()
```
